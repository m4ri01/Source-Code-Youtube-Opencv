```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(image)
```

Pada Gambar diatas, gambar yang kita load memiliki warna yang berbeda ketika ditampilkan dengan matplotlib. Hal ini dikarenakan matplotlib menggunakan channel RGB (Red, Green, Blue) sementara OpenCV meload gambar dengan kode warna BGR (Blue, Green, Red). Maka dari itu perlu dilakukan konversi warna pada gambar yang diload OpenCV

# Konversi Warna Gambar
Kita akan konversi warna gambar dari BGR ke RGB

```python
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_rgb)
```

Selain mengkonversi warna menjadi RGB, kita juga bisa membuat konversi warna grayscale pada gambar kita. Pada konversi warna grayscale, gambar kita akan diubah menjadi gambar dengan warna 1 channel yang memiliki rentang nilai dari 0 hingga 255.

```python
image_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(image_gray,cmap="gray")
```

# Memecah channel warna gambar
Selanjutnya kita akan memecah setiap channel warna pada sebuah gambar, dan membuktikan apakah benar sebuah gambar terdiri dari channel R,G, dan B

### Channel Red

```python
image_r = image_rgb.copy()
image_r[:,:,1] = 0
image_r[:,:,2] = 0
plt.imshow(image_r)
```

### Channel Green

```python
image_g = image_rgb.copy()
image_g[:,:,0] = 0
image_g[:,:,2] = 0
plt.imshow(image_g)
```

### Channel blue

```python
image_b = image_rgb.copy()
image_b[:,:,0] = 0
image_b[:,:,1] = 0
plt.imshow(image_b)
```

### Menambahkan Gambar setiap Channel
Selanjutnya kita akan coba tambahkan gambar pada setiap channel, dan membuktikan apakah akan menghasilkan gambar seperti semula?

```python
image_original = image_r+image_g+image_b
```

```python
plt.imshow(image_original)
```

```python
cv2.imwrite("image_r.png",cv2.cvtColor(image_r,cv2.COLOR_RGB2BGR))
```

```python
cv2.imwrite("image_g.png",cv2.cvtColor(image_g,cv2.COLOR_RGB2BGR))
```

```python
cv2.imwrite("image_b.png",cv2.cvtColor(image_b,cv2.COLOR_RGB2BGR))
```
